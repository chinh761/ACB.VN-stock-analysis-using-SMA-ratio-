**Nhiệm vụ**:

b) Từ hàm tính rsi, xây dựng hàm **rsi_signal(column: pandas.core.series.Series)** đánh dấu vùng quá mua (RSI > 70) và quá bán (RSI < 30), đánh dấu **"oversold"** nếu RSI < 30, "overbought" nếu RSI > 70 và **"neutral"** nếu nằm ở giữa 30 và 70 (1,5 điểm)

**Ví dụ**

df['rsi'] = rsi(df['close'], window=14)

df['rsi_signal'] = rsi_signal(df['rsi'])

print(df)

=> Kết quả:
| ticker        | timestamp        | ...                                  | rsi       | rsi_signal                    |
| ------------- | ---------------- | ------------------------------------ | --------- | ----------------------------- |
| ...           | ...              | ...                                  | ...       | ...                           |
| ACB           | 2025-04-04 00:00 | ...                                  | 26.10     | oversold                      |
| ACB           | 2025-04-08 00:00 | ...                                  | 19.42     | oversold                      |
| ...           | ...              | ...                                  | ...       | ...                           |
| ACB           | 2025-07-25 00:00 | ...                                  | 70.67     | overbought                    |
| ACB           | 2025-07-28 00:00 | ...                                  | 73.11     | overbought                    |
| ACB           | 2025-07-29 00:00 | ...                                  | 56.32     | neutral                       |
| ACB           | 2025-07-30 00:00 | ...                                  | 56.85     | neutral                       |

```python

def rsi(column: pd.Series, window: int):
    """Hàm thêm cột tính chỉ số sức mạnh tương đối (RSI) với số ngày tùy người dùng truyền vào
    *** Lưu ý: Hàm phải trả ra một pandas.Series là giá trị RSI tùy theo số ngày truyền vào từ biến window
    """
    # YOUR CODE HERE
    # Tính thay đổi trong giá
    delta = column.diff()

    # Phân chia tăng và giảm
    gain = delta.where(delta > 0, 0.0)
    loss = -delta.where(delta < 0, 0.0)

    # tính trung bình động
    avg_gain = gain.rolling(window=window, min_periods=window).mean()
    avg_loss = loss.rolling(window=window, min_periods=window).mean()

    # Tính RS và RSI
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi
df['rsi'] = rsi(df['close'], window=14)

print(df)

 
```

```python
# Không viết vào hoặc xóa cell này
```

```python
# Không viết vào hoặc xóa cell này
```

```python

def rsi_signal(column: pd.Series, window: int):
    """Hàm thêm đánh dấu vùng quá mua và quá bán
    *** Lưu ý: Hàm phải trả ra một pandas.Series đánh dấu vùng quá mua và quá bán dựa theo bảng mẫu ở trên
    """
    # YOUR CODE HERE
    return column.apply(
        lambda x: "oversold" if x < 30 else ("overbought" if x > 70 else "neutral")
    )
df['rsi'] = rsi(df['close'], window=14)

df['rsi_signal'] = rsi_signal(df['rsi'], window=14)

print(df[['timestamp', 'close', 'rsi', 'rsi_signal']])
```

    ~ SMA 20 dễ bị xuyên thủng bởi biến động do tin tức → tạo tín hiệu bán giả

```python
# YOUR CODE HERE
%pip install yfinance
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Tải dữ liệu từ yahoo finance
data_FRT = yf.download('FRT.VN', start="2024-10-29", end="2025-10-29", auto_adjust=True)

# chuẩn hoá dữ liệu
ticker = "FRT.VN"

def prepare_ohlcv(df, ticker=None):
    """
    Cần DataFrame với các cột: Open, High, Low, Close, Volume
    Dù df có MultiIndex (('Close','FRT.VN')) hay cột phẳng.
    """
    if isinstance(df.columns, pd.MultiIndex):
        if ticker is None:
            tkr = df.columns.get_level_values(-1)[0]
        else:
            tkr = ticker
        out = df.xs(tkr, axis=1, level=-1)
    else:
        out = df.copy()

    out.columns = [c.title() for c in out.columns]
    out.index = pd.to_datetime(out.index)
    out = out.sort_index()
    return out

df = prepare_ohlcv(data_FRT, ticker=ticker)
close = df["Close"]

#Tính các chỉ báo ----------
# tính thay đổi giá hằng ngày
df["daily_pct_change"] = close.pct_change() * 100.0

# RSI 14
period = 14
delta = close.diff()
gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)
avg_gain = gain.ewm(alpha=1/period, adjust=False, min_periods=period).mean()
avg_loss = loss.ewm(alpha=1/period, adjust=False, min_periods=period).mean()
rs = avg_gain / avg_loss
df["rsi14"] = 100 - 100 / (1 + rs)

# tính SMA50 & SMA200
df["sma50"] = close.rolling(window=50, min_periods=1).mean()
df["sma200"] = close.rolling(window=200, min_periods=1).mean()
print(df.tail(8)[["Close", "daily_pct_change", "rsi14", "sma50", "sma200"]])

# Vẽ biểu đồ ----------
fig, (ax1, ax2) = plt.subplots(
    2, 1, figsize=(12, 8), sharex=True, gridspec_kw={"height_ratios": [2, 1]}
)

# Format biểu đồ chính giá & SMA
ax1.plot(df.index, df["Close"], label="Close", lw=1)
ax1.plot(df.index, df["sma50"], label="SMA50")
ax1.plot(df.index, df["sma200"], label="SMA200")
ax1.set_title(f"{ticker} - Giá đóng cửa với SMA50 & SMA200")
ax1.set_ylabel("Giá")
ax1.grid(True, alpha=0.3)
ax1.legend()

# Vẽ biểu đồ RSI
ax2.plot(df.index, df["rsi14"], label="RSI-14")
ax2.axhline(70, color="red", ls="--", alpha=0.6)
ax2.axhline(30, color="green", ls="--", alpha=0.6)
ax2.set_ylim(0, 100)
ax2.set_title("RSI-14 (70=quá mua, 30=quá bán)")
ax2.set_ylabel("RSI")
ax2.grid(True, alpha=0.3)

# Sửa lại trục thời gian
locator = mdates.AutoDateLocator()
ax2.xaxis.set_major_locator(locator)
ax2.xaxis.set_major_formatter(mdates.ConciseDateFormatter(locator))

plt.tight_layout()
plt.show()
print('Bối cảnh:\n Trong giai đoạn Tổng thống Trump áp đặt thuế quan, thị trường chứng khoán Việt Nam đã chịu tác động mạnh. \n Tâm lý nhà đầu tư bi quan, dẫn đến việc rút vốn và giá cổ phiếu giảm sâu. Giai đoạn này, chỉ báo RSI liên tục ở dưới ngưỡng 30 (quá bán), trong khi SMA50 cắt xuống dưới SMA200 (tạo ra Death Cross), phản ánh một xu hướng giảm giá mạnh mẽ.')
print('Ưu điểm sử dụng các chỉ báo:')
print('RSI phản ứng nhanh chóng: Khi giá giảm mạnh và RSI duy trì dưới 30, nhà đầu tư có thể nhận biết sớm tình trạng thị trường đã quá bán và có khả năng hồi phục kỹ thuật trong ngắn hạn.')
print('SMA giúp xác định xu hướng dài hạn:  SMA200 vẫn ổn định trong khi SMA50 và giá giảm mạnh -> xu hướng suy yếu đang lan rộng, nhà đầu tư cần cảnh giác với rủi ro giảm giá kéo dài.')
print('--> Khi RSI ở vùng quá bán và giá cách xa các đường SMA, có thể dự báo một đợt bán tháo mạnh --> tạo ra cơ hội mua vào tiềm năng để chốt lời khi giá hồi phục.')
print('Nhược điểm:')
print('Tín hiệu quá bán kéo dài: Khi 1 cú sốc vĩ mô lớn xảy ra, như việc áp thuế, RSI duy trì ở vùng quá bán trong thời gian dài.\n -> khiến tín hiệu “bắt đáy” của RSI không hiệu quả, vì nhà đầu tư mua vào sẽ tiếp tục chịu lỗ.')
print('Tín hiệu giả: Khi thị trường biến động mạnh bởi các tin tức, giá có thể cắt các đường SMA rồi quay lại.\n Điều này tạo ra tín hiệu bán giả, khiến nhà đầu tư bán ra sớm.\n Tương tự, RSI tạm thời bật lên khỏi vùng 30, nhưng không đủ mạnh để đảo chiều xu hướng-> nhà đầu lầm tưởng rằng thị trường hồi phục dù xu hướng giảm tiếp tục.')
```
